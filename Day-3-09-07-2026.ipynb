```python
# Green Skilling & Sustainability Dataset Analysis
import pandas as pd
df = pd.read_csv("GreenSkilling_Sustainability_Dataset.csv")
```

```python
print("\nStatistical Summary")
print(df.describe())
```

```python
# Generalizations and Analysis

# 1. Country
# Represents different countries.

# 2. Solar_Energy_%
# Higher solar percentage usually indicates more renewable energy usage.

# 3. Wind_Energy_%
# Shows the contribution of wind energy.

# 4. Green_Skilling_Training
# Number of people trained in green skills.

# 5. Green_Jobs
# Number of green employment opportunities.

# 6. CO2_Emissions_tonnes_per_capita
# Lower values are better for sustainability.

# 7. Recycling_Rate_%
# Higher recycling rate generally indicates better waste management.
```

```python
# Relation Between Columns

# Solar Energy ↑
#      ↓
# CO2 Emissions ↓

# Wind Energy ↑
#      ↓
# CO2 Emissions ↓

# Green Skill Training ↑
#      ↓
# Green Jobs ↑

# Recycling Rate ↑
#      ↓
# Better Environmental Performance

# Countries with higher renewable energy generally tend to have
# lower CO2 emissions.
```

```python
# Check Null Values

print("\nNull Values")
print(df.isnull().sum())
```

```python
# Numerical columns are filled with their mean values.

df["Solar_Energy_%"] = df["Solar_Energy_%"].fillna(df["Solar_Energy_%"].mean())

df["Green_Skilling_Training"] = df["Green_Skilling_Training"].fillna(
    df["Green_Skilling_Training"].mean()
)

print("\nNull Values After Handling")
print(df.isnull().sum())
```

```python
# Correlation

print("\nCorrelation Matrix")
print(df.corr(numeric_only=True))
```
